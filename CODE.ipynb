# 1.Importing Libraries

```python
#Q1.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(plt)
```

## 2.Data Loading

The datasets `male.csv` and `female.csv` were successfully loaded into Pandas DataFrames using `pd.read_csv`. These DataFrames were then converted into NumPy matrices named `male` and `female` using the `.values` attribute. Each matrix contains seven columns representing weight (kg), standing height (cm), upper arm length (cm), upper leg length (cm), arm circumference (cm), hip circumference (cm), and waist circumference (cm), with the number of rows corresponding to the number of participants in each dataset.

- **Male Matrix Shape**: The `male` matrix shape was verified, indicating the number of participants and the 7 features.
- **Female Matrix Shape**: The `female` matrix shape was similarly confirmed, matching the expected structure.

This step ensures the data is properly structured for subsequent analysis.

```python
#Q2.
male_df = pd.read_csv('male.csv')
female_df = pd.read_csv('female.csv')
male = male_df.values
female = female_df.values
# To Verify shapes
print("Male matrix shape:", male.shape)
print("Female matrix shape:", female.shape)
```

# 3.Histogram Comparison of Male and Female Weights
Two histograms were plotted using `matplotlib.pyplot.subplot()` â€” one for female weights (top) and one for male weights (bottom). Both plots share the same x-axis limits to make comparisons easier.

The female weight histogram uses a sky-blue color with black borders, showing how female weights are distributed.

The male weight histogram uses a light-green color with dark green edges, representing the distribution of male weights.

The `plt.xlim()` function ensures both histograms cover the same range of values, allowing direct visual comparison.

Finally, `plt.tight_layout()` was used to prevent overlapping between the two plots.

```python
#Q3.
# Replace 'Weight' with the actual column name if different
female_weights = female[:, 0]
male_weights = male[:, 0]

# Determine common x-axis limits
xlim_min = min(np.min(female_weights), np.min(male_weights))
xlim_max = max(np.max(female_weights), np.max(male_weights))

#create  subplots
plt.figure(figsize=(8, 8))

#Top subplot - Female
plt.subplot(2, 1, 1)
plt.hist(female_weights, bins=20, color='skyblue', edgecolor='black', alpha=0.8)
plt.title('Histogram of Female Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.xlim(xlim_min - 5, xlim_max + 5)

#Bottom subplot - Male
plt.subplot(2, 1, 2)
plt.hist(male_weights, bins=20, color='lightgreen', edgecolor='darkgreen', alpha=0.8)
plt.title('Histogram of Male Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.xlim(xlim_min - 5, xlim_max + 5)

plt.tight_layout() #for space between two histograms
plt.show()
```

# 4.Box-and-Whisker Plot for Male and Female Weights

In this visualization, a box-and-whisker plot was created using `matplotlib.pyplot.boxplot()` to compare male and female weight distributions side by side.

The figure size was enlarged using `figsize=(20, 16)` for better visibility.

Two boxes were plotted â€” light pink for females and default white for males â€” with red median lines to highlight central tendency.

The whiskers and caps were drawn in black to clearly show the spread of data, while blue circular markers indicated outliers.

A grid along the y-axis (`plt.grid()`) made it easier to interpret weight differences.

## Summary
The boxplot effectively displays the differences in median, range, and variability between male and female weights. It also highlights potential outliers and provides a clear comparison of how weight values are spread within each group.

```python
#Q4.
# Create a boxplot for comparison
plt.figure(figsize=(20, 16))
plt.boxplot([female_weights, male_weights],
            labels=['Female', 'Male'],
            patch_artist=True,   # enable color fill
            boxprops=dict(facecolor='lightpink', color='black'),
            medianprops=dict(color='red', linewidth=2),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            flierprops=dict(marker='o', color='blue', alpha=0.5))

plt.title('Comparison of Male and Female Weights')
plt.ylabel('Weight (kg)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 6.Adding BMI to Female Data

## Formula Used


$$BMI = \frac{\text{Weight (kg)}}{\text{Height (m)}^2}$$


## Process
- Computed BMI for all female participants.
- Added BMI as a new (8th) column to the female dataset.
- Verified the new matrix shape using: `female.shape`

## Result
- Female matrix now includes an extra column representing each participantâ€™s BMI.

```python
#Q6.
# Calculate BMI
bmi = female[:, 0] / ((female[:, 1] / 100) ** 2)

# Add BMI as a new column
female = np.column_stack((female, bmi))

# Verify the new shape
print("\nFemale matrix shape after adding BMI:", female.shape)
```


# 7. Z-Score Normalization (Standardization)

- Each value in the `female` matrix is standardized using:  
$$
z = \frac{x - \bar{x}}{s}
$$
where:  
- \(x\) = original value  
- \(\bar{x}\) = mean of the feature  
- \(s\) = standard deviation of the feature  
*  Centers the data to mean 0 and scales to standard deviation 1.  
*  Resulting matrix `zfemale` has the same shape as `female` but is normalized for analysis.


```python
#Q7.Create standardized zfemale matrix
zfemale = (female - np.mean(female, axis=0)) / np.std(female, axis=0)
print("zfemale matrix shape:", zfemale.shape)
```

```python
#Q8.
# Scatterplot matrix and correlations
# columns for pairplot: height (1), weight (0), waist (6), hip (5), BMI (7)
columns = ['Weight', 'Height', 'Waist', 'Hip', 'BMI']
zfemale_subset = zfemale[:, [0, 1, 6, 5, 7]]

# Convert to DataFrame for pairplot and correlation
zfemale_df = pd.DataFrame(zfemale_subset, columns=columns)

# Draw pairplot
sns.pairplot(zfemale_df)
plt.show()

# Compute correlations
pearson_corr = zfemale_df.corr(method='pearson')
spearman_corr = zfemale_df.corr(method='spearman')

print("----------------------------------------------------------")
print("Pearson Correlation Coefficients:\n", pearson_corr)
print("----------------------------------------------------------")
print("Spearman Correlation Coefficients:\n", spearman_corr)
print("----------------------------------------------------------")
```

# 9.Waist-to-Height and Waist-to-Hip Ratios

**Purpose:**  
Compute and append two ratiosâ€”**waist/height** and **waist/hip**â€”for both female and male datasets.

**Steps:**  
1. Print initial shapes of `female` and `male` matrices.  
2. Calculation:  
   
  * Waist-to-Height Ratio:  
  $$
  \text{waist\_height} = \frac{\text{Waist}}{\text{Height}}
  $$

  * Waist-to-Hip Ratio:  
  $$
  \text{waist\_hip} = \frac{\text{Waist}}{\text{Hip}}
  $$
  
3. Append these ratios as new columns to the respective matrices.  
4. Print updated shapes to confirm new columns added.

**Result:**  
Both matrices now have **two additional columns** representing the computed ratios.


```python
#Q9
print("Current female matrix shape:", female.shape)
print("Current male matrix shape:", male.shape)

# Calculate and append ratios for female
female_waist_height = female[:, 6] / female[:, 1]  # Waist / Height
female_waist_hip = female[:, 6] / female[:, 5]    # Waist / Hip
female = np.column_stack((female, female_waist_height, female_waist_hip))

# Calculate and append ratios for male
male_waist_height = male[:, 6] / male[:, 1]       # Waist / Height
male_waist_hip = male[:, 6] / male[:, 5]          # Waist / Hip
male = np.column_stack((male, male_waist_height, male_waist_hip))

# Verify updated shapes
print("----------------------------------------")
print("Updated female matrix shape:", female.shape)
print("Updated male matrix shape:", male.shape)
```

# 11.Advantages and Disadvantages of Metrics
1. Body Mass Index (BMI)
------------------------
Advantages:
- Simple and quick to calculate using just weight and height.
- Useful for large-scale health screenings and population studies.
- Provides a general indication of whether an individualâ€™s weight is within a healthy range.

Disadvantages:
- Cannot distinguish between fat and muscle mass, so muscular individuals may appear overweight.
- Does not indicate fat distribution (e.g., belly fat vs. hip fat).
- May not accurately reflect health risks for all body types.

2. Waist-to-Height Ratio (WHtR)
-------------------------------
Advantages:
- Offers a better measure of abdominal fat, which is strongly linked to heart disease and diabetes.
- Adjusts for height, making it more personalized than BMI.
- Helps identify health risks in people of different body sizes.

Disadvantages:
- Still doesnâ€™t account for muscle mass, possibly overestimating risk for athletic individuals.
- Less widely used than BMI, so it has less clinical and research support.

3. Waist-to-Hip Ratio (WHR)
------------------------------
Advantages:
- Indicates fat distribution pattern (e.g., apple-shaped vs. pear-shaped body).
- Helps assess risks for cardiovascular diseases and metabolic conditions.
- Highlights central obesity more effectively than BMI.

Disadvantages:
- Can be influenced by bone structure or body frame, reducing accuracy.
- Less practical for large-scale or quick assessments.

Measurement errors (e.g., incorrect tape placement) can affect reliability.

# 12.BMI Extremes Analysis for Female Individuals

This analysis identifies and displays the standardized body measurements for female individuals at the extreme ends of the Body Mass Index (BMI) spectrum.

## Process
1. The female individuals are sorted based on their BMI values.
2. The five individuals with the **lowest BMI** and the five with the **highest BMI** are isolated.
3. The complete set of standardized measurements (`zfemale`) for these two distinct groups is then printed for comparison.

## Purpose
The goal is to compare the physiological profiles of individuals with the most divergent BMIs. By examining their standardized measurements, one can observe which other body features (such as height, waist circumference, etc.) consistently correlate with very low and very high BMI values. This is useful for understanding the physical characteristics associated with **underweight** and **overweight** individuals in the dataset.


```python
#Q12
# Sort indices based on BMI values
bmi_sorted_indices = np.argsort(female[:, 7])  # Column 7 is BMI in female matrix

# Get indices for 5 lowest and 5 highest BMI
lowest_5_indices = bmi_sorted_indices[:5]
highest_5_indices = bmi_sorted_indices[-5:]

# Print standardized measurements
print("Standardized measurements for 5 persons with the lowest BMI:")
print(zfemale[lowest_5_indices])
print("\nStandardized measurements for 5 persons with the highest BMI:")
print(zfemale[highest_5_indices])
```
